```python
import pandas as pd
import tkinter as tk 
from tkinter import messagebox

# Load the data
employee_data = pd.read_csv("GIG-logistics.csv")

def welcomeMessage(username, dept_name, colleagues):
    
    window = tk.Toplevel(root)
    window.title("Admin Box")
    window.geometry("500x300")

    tk.Label(window, text=f"{username.upper()} is a member of {dept_name.upper()}").pack()
    tk.Label(window, text="Other members in the department:").pack()

    colleagues = "\n".join(colleagues) if colleagues else "No other members."
    
    tk.Label(window, text=colleagues).pack()

def verify_employee():
    
    e_name = employee_entry.get().strip()
    e_department = department_entry.get().strip()

    match = employee_data[(employee_data['Name'].str.lower() == e_name.lower()) & 
                          (employee_data['Department'].str.lower() == e_department.lower())]

    if not match.empty:
      
        others = employee_data[(employee_data['Department'].str.lower() == e_department.lower()) & 
                              (employee_data['Name'].str.lower() != e_name.lower())]
        
        teammate_names = others['Name'].tolist()
        
        welcomeMessage(e_name, e_department, teammate_names)
    else:
        messagebox.showerror("Login", "No employee found with that name and department")

# Create main window 
root = tk.Tk()
root.title("GIG-logistics Employee verification page")
root.geometry("300x200")

# e_name label and entry 
tk.Label(root, text="Surname: ").pack()
employee_entry = tk.Entry(root)
employee_entry.pack()

# Department label and entry 
tk.Label(root, text="Department: ").pack()
department_entry = tk.Entry(root)
department_entry.pack()

# verify button
submit_button = tk.Button(root, text="Submit", command=verify_employee)
submit_button.pack()

# Run the mainloop
root.mainloop()
```
