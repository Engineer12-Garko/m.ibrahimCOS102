```python
import tkinter as tk
from tkinter import messagebox

class DeliveryApp:
    def __init__(self, root):

        # creating the window
        self.root = root
        self.root.title("Delivery Price Calculator")
        self.root.geometry('400x500')

        #Setting the destination Input
        self.label1 = tk.Label(root, text="Enter Location (PAU or Epe):")
        self.label1.pack(pady=5)
        self.location_entry = tk.Entry(root)
        self.location_entry.pack(pady=5)

        # Setting the Weight Input
        self.label2 = tk.Label(root, text="Enter Weight (kg):")
        self.label2.pack()
        self.weight_entry = tk.Entry(root)
        self.weight_entry.pack(pady=5)

        # creating a button to click
        self.calc_button = tk.Button(root, text="Get Price", command=self.check_price)
        self.calc_button.pack(pady=5)

    def check_price(self):
        
        loc = self.location_entry.get().lower()
        weight = float(self.weight_entry.get())
        
        # The price logic
        price = 0

        if loc == "pau":
            if weight >= 10:
                price = 2000
            else:
                price = 1500
        
        elif loc == "epe":
            if weight >= 10:
                price = 5000
            else:
                price = 4000
        
        
        messagebox.showinfo("Result", "Total Price: N" + str(price))

# Main window
root_window = tk.Tk()
app = DeliveryApp(root_window)
root_window.mainloop()
```
