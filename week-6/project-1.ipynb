```python
import tkinter as tk
from tkinter import messagebox
import random as rd

class Logistics_app:
    def __init__(self, root):
        self.root = root
        self.root.title("Logistics App")
        self.root.geometry("400x300")

        self.employees = [
            "Mary Evans", "Eyo Ishan", "Durojaiye Dare", "Adams Ali", 
            "Andrew Ugwu", "Stella Mankinde", "Jane Akibo", "Ago James", 
            "Michell Taiwo", "Abraham Jones", "Nicole Anide", "Kosi Korso", 
            "Adele Martins", "Emmanuel Ojo", "Ajayi Fatima"
        ]
        
        self.tasks = ["Loading", "Transporting", "Reviewing Orders", "Customer Service", "Delivering Items"]

        self.label = tk.Label(self.root, text="Employee Name:")
        self.label.pack(pady=10)

        self.entry = tk.Entry(self.root)
        self.entry.pack(pady=5)

        self.button = tk.Button(self.root, text="Verify & Assign Task", command=self.check_employee)
        self.button.pack(pady=20)

    def check_employee(self):
     
        name = self.entry.get().strip()

        task_num = rd.randint(1, 5) 

        if name in self.employees:
            if task_num == 1:
                task_name = self.tasks[0]
            elif task_num == 2: 
                task_name = self.tasks[1]
            elif task_num == 3:
                task_name = self.tasks[2]
            elif task_num == 4:
                task_name = self.tasks[3]
            elif task_num == 5:
                task_name = self.tasks[4]
            
            messagebox.showinfo("Confirmed", f"Welcome {name}!\nAttendance marked.\nTask: {task_name}")
        
        else:
           
            messagebox.showerror("Not confirmed", "Sorry, name not found!")


window = tk.Tk()
app = Logistics_app(window)
window.mainloop()
```
