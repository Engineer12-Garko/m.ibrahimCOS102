```python
# (a) simple interest 
print('Simple interest Calculation program')
P = float(input("Enter the Principal (P): "))
R = float(input("Enter the Rate (R): "))
T = float(input("Enter time Time (T): "))

# Applying the formula from the flowchart
A = P * (1 + (R / 100) * T)

print("The total amount (A) is:", A)
```

```python
# (b) Compound interest 
print('Compound interest calculation progarm')
P = float(input("Enter Principal (P): "))
R = float(input("Enter Annual Interest Rate (R): "))
T = float(input("Enter Time in years (T): "))
N = float(input("Enter number of times interest is compounded per year (N): "))

# Applying the formula in the flowchart
A = P * (1 + (R / N))**(N * T)

print("The compound amount (A) is:", A)
```

```python
# (c) Annuity plan
print('Annuity Plan calculation program')
PMT = float(input("Enter Periodic Payment (PMT): "))
R = float(input("Enter Annual Interest Rate (R): "))
T = float(input("Enter Time in years (T): "))
N = float(input("Enter number of payments per year (N): "))

# Applying the formula in the flowchart
num = (1 + (R / N))**(N * T) - 1
denom = R / N
A = PMT * (num / denom)

print("The future value of the annuity (A) is:", A)
```
