```python
import numpy as np
print("Program to find roots of Quadratic, Cubic, and Quartic equations")
print("1. Quadratic")
print("2. Cubic")
print("3. Quartic")

equation = input("\nSelect the Equation (1, 2, or 3): ") 

if equation == '1':
    print("\nEnter coefficients: ")
    a = float(input("a: "))
    b = float(input("b: "))
    c = float(input("c: "))
    coeffs = [a, b, c]

elif equation == '2':
    print("\nEnter coefficients: ")
    a = float(input("a: "))
    b = float(input("b: "))
    c = float(input("c: "))
    d = float(input("d: "))
    coeffs = [a, b, c, d]
    
elif equation == '3':
    print("\nEnter coefficients: ")
    a = float(input("a: "))
    b = float(input("b: "))
    c = float(input("c: "))
    d = float(input("d: "))
    e = float(input("e: "))
    coeffs = [a, b, c, d, e]

else:
    print("Invalid selection.")

# using numpy to find roots
roots = np.roots(coeffs)

print("\nRoots: ")
for i in roots:
    result = round(i, 2)
    print("The root is:", result)
```
