```python
# Practice-1/cell-2
# Input from user
m = float(input("Enter the mass in kilograms: "))

#  constant value for the speed of light in m/s
c = 299792458

# calculating energy Eintein's equation
energy = m * c ** 2

# Displaying the result
print(f"The energy equyivalent to {m} kg of mass is {energy}joules.")

```

```python
# Practice-1/cell-3

list =['Anaconda', 786, 2.23, 'Jupyter', 70.2]
shortlist = [321, 'Python']

print(list)                 # Prints complete list 
print(list[0])              # Print first element pof the list 
print(list[1:3])            # Prints elements starting from 2nd till 3rd
print(list[2:])             # Prints elements starting from 3rd element 
print(shortlist * 2)        # Prints shortlist two times 
print(list + shortlist)     # Prints concateneted lists
```

```python
# Practice-1/cell-4
tuple = ("Ekiti", 750, "Oshogbo", 250, "Akure", 500)
s_tuple = ("Abekuta", 300, "Ogbomoso")

# Prints the complete tuple
print(tuple)

# Prints last element of the tuple
print(tuple[-1])

# prints elements of tuple starting from 2nd element till 3rd
print(tuple[1:4])

# Prints elements of tuple starting from 3rd element
print(tuple[3:])

# Prints the contents of the tuple twice
print(s_tuple * 2)

# Prints concatenated tuples
print(tuple + s_tuple)
```
