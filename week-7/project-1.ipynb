```python
import tkinter as tk
from tkinter import messagebox


class Zenith:
    def __init__(self, name):
        self.name = name

    def mutual_services(self):
        return ["Lines of credit", "Investment management and accounts", "Insurance"]

    def unique_services(self):
        return []



class RetailBanking(Zenith):
    def unique_services(self):
        return [ "Retirement and education accounts", "Loans and mortgages", "Checking and savings"]


class GlobalBanking(Zenith):
    def unique_services(self):
        return ["Multi-currency management services and products",
            "Foreign currency accounts",
            "Foreign currency credit cards",
            "Transborder advisory services",
            "Liquidity management"]


class CommercialBanking(Zenith):
    def unique_services(self):
        return ["Advisory services"]



def show_services():
    name = name_entry.get()
    division = division_entry.get().lower()

    if name == "" or division == "":
        messagebox.showerror("Error", "Please enter both name and division")
        return

    if division == "retail":
        user = RetailBanking(name)
    elif division == "global":
        user = GlobalBanking(name)
    elif division == "commercial":
        user = CommercialBanking(name)
    else:
        messagebox.showerror("Error", "Division must be Retail, Global, or Commercial")
        return

    # get services
    mutual = user.mutual_services()
    unique = user.unique_services()

    # display result in new window
    window = tk.Toplevel(root)
    window.title("Services Display")
    window.geometry("400x400")

    tk.Label(window, text=f"Welcome {name}").pack()

    tk.Label(window, text="\nMutual Services:").pack()
    for service in mutual:
        tk.Label(window, text=f"- {service}").pack()

    tk.Label(window, text="\nUnique Services:").pack()
    for service in unique:
        tk.Label(window, text=f"- {service}").pack()




root = tk.Tk()
root.title("Zenith Bank System")
root.geometry("400x250")

# Name input
name_label = tk.Label(root, text="Employee Name:")
name_label.pack()
name_entry = tk.Entry(root)
name_entry.pack()

# Division input
division_label = tk.Label(root, text="Division (Retail / Global / Commercial):")
division_label.pack()
division_entry = tk.Entry(root)
division_entry.pack()

# Submit button
submit_btn = tk.Button(root, text="Submit", command=show_services)
submit_btn.pack(pady=10)

root.mainloop()
```
